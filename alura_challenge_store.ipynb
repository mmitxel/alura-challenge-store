# Gráficos

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

## Organización de los datos para graficarlos

```python
# Organización de los datos para graficarlos
tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

tiendas = {
    'Tienda 1': tienda1,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

resultados = pd.DataFrame({
    'Facturación': [t['Precio'].sum() for t in tiendas.values()],
    'Calificación': [t['Calificación'].mean() for t in tiendas.values()],
    'Costo de envío': [t['Costo de envío'].mean() for t in tiendas.values()],
    'Categorías de Productos': [t.groupby('Categoría del Producto').size() for t in tiendas.values()]
}, index=tiendas.keys())
```

## Comparativa de facturación en las cuatro tiendas

```python
ventas = resultados['Facturación'].plot(kind='bar', ylabel='Facturación ($)',
                                        figsize=(8,6), legend=False,  color='steelblue')

plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x*1e-6:.1f}M'))
#plt.grid(axis='y', linestyle=':', alpha=0.8)
ventas.bar_label(
    ventas.containers[0],
    fmt=lambda x: f'{x*1e-6:.1f}M',
    padding=-15,
    color='lightcyan',
    weight='demi')
ventas.locator_params(axis='y', nbins=15)
plt.xticks(rotation=0)

plt.show()
```

```python
s = resultados['Facturación']
base = s.min()
pct = (s / base - 1) * 100

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(
    s.index,
    pct,
    color='coral',
    marker='o',
    linewidth=2
)

for x, y in zip(s.index, pct):
    ax.text(x, y + 0.5, f'{y:.1f}%', ha='center', va='bottom', color='indianred', fontsize=10)

ax.set_ylim(0, 13)
ax.set_title('Diferencia relativa de facturación vs. Tienda 4')
ax.set_ylabel('Porcentaje sobre Tienda 4')
ax.grid(axis='y', linestyle=':', alpha=0.6)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Comparativa de costos de envío

```python
costos_envio = resultados['Costo de envío'].plot(kind='bar', ylabel='Costo de envío promedio (pesos colombianos)' , figsize=(8,6), legend=False,  color='mediumseagreen')
plt.grid(axis='y', linestyle='--', alpha=1, color='green')

plt.xticks(rotation=0)

plt.show()
```

## Comparativa de categorías de productos vendidos

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch # Rectángulos de color en categrías

series_por_tienda = resultados["Categorías de Productos"]
categorias = list(series_por_tienda.iloc[0].index)

palette_soft_harmony = [
    "#E57373",  # soft brick red
    "#F4A261",  # muted orange, earthy and warm (like terracotta)
    "skyblue",  # clear sky blue
    "#A8E36B",
    "#BA68C8",  # soft purple
    "lightseagreen",  # teal
    "#F4D35E",  # mellow yellow (natural, not bright neon)
    "#A1887F"   # taupe brown / neutral
]

color_map = {cat: palette_soft_harmony[i] for i, cat in enumerate(categorias)}

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.flatten()

for i, tienda in enumerate(series_por_tienda.index):
    s = series_por_tienda[tienda].sort_values()

    axs[i].pie(
        s.values,
        labels=None,
        autopct='%1.1f%%',
        startangle=90,
        colors=[color_map[c] for c in s.index]
    )

    axs[i].set_title(tienda)
    axs[i].axis('equal')


categorias = series_por_tienda.iloc[0].index
handles = [Patch(facecolor=color_map[c], edgecolor='none') for c in categorias]

fig.legend(
    handles,
    categorias,
    loc='center',
    ncol=1,
    bbox_to_anchor=(0.5, 0.47)
)

plt.suptitle('Comparativa de categorías de productos vendidos', fontsize=20, weight='demi')
plt.tight_layout()
plt.show()
```

## Comparativa de calificaciones de clientes

```python
calificaciones = resultados['Calificación'].plot(kind='barh',
                                                 xlabel='Calificación promedio',
                                                 figsize=(6, 6),
                                                 legend=False,
                                                 color='goldenrod')
plt.grid(axis='x', linestyle=':', alpha=0.8)
calificaciones.set_xlim(0, 5)
plt.yticks(rotation=90 )
calificaciones.bar_label(calificaciones.containers[0],
                         fmt='%.2f',
                         padding=10,
                         color='darkgoldenrod',
                         weight='demi')
for label in calificaciones.get_yticklabels():
    label.set_va('center')  # moves the text baseline upward

plt.show()
```

# Informe Final

Se realizó un análisis de las métricas en las cuatro tiendas para encontrar cuál de ellas sería más conveniente vender. Se analizaron tanto las variables de ventas y satisfacción de clientes así como aquellas que impactan estrategias de ventas.

Se descubrió que la **tienda número 4** muestra una clara desventaja respecto a las demás por su menor rentabilidad a pesar de contar con ventajas estratégicas respecto a las demás:

Primero, su **facturación** total es significativemente menor a la de las otras tiendas, siendo solamente de 1 308.4 millones de pesos colombianos (COP), contra 1 150.9, 1 116.3, y 1 098.0 millones COP de las otras tres (fig. 1), representando una diferencia que va del 6% al 11% (fig. 2).

En segundo lugar, se encontró que la tienda número 4 cuenta con una ventaja en sus **costos de envío** respecto a las demás (fig 3), siendo estos significativamente más bajos en promedio, lo cual no se ha reflejado en mayores ventas como se podría esperar.

Al analizar las categorías de productos vendidos, se encontraron diferencias en las tiendas 3 y 4, respecto a las otras dos. En la número tres, las ventas de **libros** son mayores a las de **instrumentos musicales**, y en la número cuatro, las ventas de **artículos para el hogar** son mayores a las de las dos categorías anteriomente mencionadas, lo cual difiere con las otras tres tiendas. Más significativamente, las ventas en la categoría de **deportes y diversión** en la tienda 4 superan a las de **electrodomésticos**, a diferencia de las otras tres tiendas también (fig. 4). Esto significa que cualquier estrategia de marketing como descuentos en productos o campañas promocionales tendrán que tomar en cuenta las diferencias de la tienda 4, lo cual puede complicar su implementación y reducir su efectividad.

Por último, en cuanto a la calificación que los clientes dan a las tiendas, no se encontró una diferencia significativa en ninguna de ellas, ya que todas se encuentran entre el 3.98 y el 4.05 (fig. 5), por lo que esto no afectó la decisión en un sentido u otro.

En conclusión, se recomienda la venta de la tienda número 4 debido a su menor rentabilidad y desventaja competitiva.
